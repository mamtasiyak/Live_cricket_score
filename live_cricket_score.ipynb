```python
from pycricbuzz import Cricbuzz
```

```python
cric=Cricbuzz()
```

```python
match=cric.matches()
scorecard=cric.scorecard('20244')
scorecard['scorecard']
```

```python
match=cric.matchinfo('20244')
match
```

```python
match['team2']['name'] 
```

```python
match['team2']['squad']
```

```python
print('series ->',match['srs'])
print('Match ->',match['team2']['name'],' Vs ',match['team1']['name'],'\n\n')
print(match['team2']['name'],'-',scorecard['scorecard'][0]['runs'],'/',scorecard['scorecard'][0]['wickets'],'      ',match['team1']['name'],'-',scorecard['scorecard'][1]['runs'],'/',scorecard['scorecard'][1]['wickets'])
print('Status ->',match['status'],'\n\n')
print('Team ->',match['team1']['name'])
print('Squad ->',match['team1']['squad'],'\n\n')
print('Team ->',match['team2']['name'])
print('Squad ->',match['team2']['squad'],'\n\n')

print('Team India Batting')
for i in range(len(scorecard['scorecard'][0]['batcard'])):
    print(scorecard['scorecard'][0]['batcard'][i]['name'],'-',scorecard['scorecard'][0]['batcard'][i]['runs'])

print('\n\nTeam South Africa Batting')
for i in range(len(scorecard['scorecard'][1]['batcard'])):
    print(scorecard['scorecard'][1]['batcard'][i]['name'],'-',scorecard['scorecard'][1]['batcard'][i]['runs'])
    
print('\n\nTeam ',match['team2']['name'],' Bowling')
for i in range(len(scorecard['scorecard'][1]['bowlcard'])):
    print('Name-',scorecard['scorecard'][1]['bowlcard'][i]['name'],'  runs-',scorecard['scorecard'][1]['bowlcard'][i]['runs'],'  overs-',scorecard['scorecard'][1]['bowlcard'][i]['overs'],'  wickets-',scorecard['scorecard'][1]['bowlcard'][i]['wickets'])
    
print('\n\nTeam ',match['team1']['name'],' Bowling')
for i in range(len(scorecard['scorecard'][0]['bowlcard'])):
    print('Name-',scorecard['scorecard'][0]['bowlcard'][i]['name'],'  runs-',scorecard['scorecard'][0]['bowlcard'][i]['runs'],'  overs-',scorecard['scorecard'][0]['bowlcard'][i]['overs'],'  wickets-',scorecard['scorecard'][0]['bowlcard'][i]['wickets'])    
```
